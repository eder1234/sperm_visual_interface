```python
import pandas as pd

file_path = "traj-partial.txt"

df = pd.read_csv(file_path, names=["name", "date", "quantity", "exposure", "tracked_id", "x", "y"], engine='python') 
df
```

```python
traj_df = pd.DataFrame({"name":[], "date":[], "quantity":[], "exposure":[], "tracked_id":[], "traj":[]})

df1 = df.groupby('name')
for k1, v1 in df1:
    df2 = df1.get_group(k1)
    df3 = df2.groupby('date')
    for k3,v3 in df3:
        df4 = df3.get_group(k3)
        df5 = df4.groupby('quantity')
        for k5,v5 in df5:
            df6= df5.get_group(k5)
            df7 = df6.groupby('exposure') 
            for k7,v7 in df7:
                df8 = df7.get_group(k7)
                df9 = df8.groupby('tracked_id')
                for k9,v9 in df9:
                    df10 = df9.get_group(k9)
                    x = df10['x'].to_numpy()
                    y = df10['y'].to_numpy()
                    temp_df = pd.DataFrame({"name":[k1], "date":[k3], "quantity":[k5], "exposure":[k7], "tracked_id":[k9], "traj":[[x,y]]})
                    traj_df = pd.concat([temp_df, traj_df], ignore_index=True)
```

```python
traj0_list = master0_df['traj'].values.tolist()
traj30_list = master30_df['traj'].values.tolist()
```

```python
import numpy as np

def center_of_traj(x, y):
    max_x = np.amax(x)
    min_x = np.amin(x)
    max_y = np.amax(y)
    min_y = np.amin(y)
    x_c = np.mean([max_x, min_x])
    y_c = np.mean([max_y, min_y])
    return x_c, y_c 
```

```python
import matplotlib.pyplot as plt
from PIL import Image
from tqdm import tqdm
import numpy as np

window_size = 60
for index, traj in tqdm(enumerate(traj0_list)):
    list_x, list_y = zip(traj)
    x = np.asarray(list_x).squeeze()
    y = np.asarray(list_y).squeeze()
    c = range(len(x)) #arange numpy
    x_c, y_c = center_of_traj(x, y)
    plt.figure(figsize=[2,2])
    plt.plot(x, y, 'k')
    plt.xlim([x_c - window_size/2, x_c + window_size/2])
    plt.ylim([y_c - window_size/2, y_c + window_size/2])
    plt.gca().set_aspect('equal', adjustable='box')
    plt.axis('off')
    new_id = '{:05d}'.format(index)
    plt.savefig(f'img_0_05052022/{new_id}.png')
    plt.close()
    
for index, traj in tqdm(enumerate(traj30_list)):
    list_x, list_y = zip(traj)
    x = np.asarray(list_x).squeeze()
    y = np.asarray(list_y).squeeze()
    c = range(len(x)) #arange numpy
    x_c, y_c = center_of_traj(x, y)
    plt.figure(figsize=[2,2])
    plt.plot(x, y, 'k')
    plt.xlim([x_c - window_size/2, x_c + window_size/2])
    plt.ylim([y_c - window_size/2, y_c + window_size/2])
    plt.gca().set_aspect('equal', adjustable='box')
    plt.axis('off')
    new_id = '{:05d}'.format(index)
    plt.savefig(f'img_30_05052022/{new_id}.png')
    plt.close()
```

# Load the images

```python
import pathlib

traj0_img_list =  sorted(str(p) for p in pathlib.Path("img_0_05052022/").glob("*.png"))
traj30_img_list =  sorted(str(p) for p in pathlib.Path("img_30_05052022/").glob("*.png"))

print(len(traj0_img_list))
print(len(traj30_img_list))
```

```python
import numpy as np
from PIL import Image
from numpy import asarray

list0_arrays = []
for elem in traj0_img_list:
    img = Image.open(elem).convert('L')
    numpydata = asarray(img).flatten()
    list0_arrays.append(numpydata)
    
train0_np = np.asarray(list0_arrays)
train0_tensor = train0_np/255

list30_arrays = []
for elem in traj30_img_list:
    img = Image.open(elem).convert('L')
    numpydata = asarray(img).flatten()
    list30_arrays.append(numpydata)
    
train30_np = np.asarray(list30_arrays)
train30_tensor = train30_np/255
```

## Generate the labels

```python
from sklearn.cluster import DBSCAN

clustering0 = DBSCAN(eps=0.5, min_samples=2).fit(train0_tensor)
y0_labels = clustering0.labels_
n0_clusters = np.amax(y0_labels)

print(y0_labels)
print(n0_clusters)
```

```python
clustering30 = DBSCAN(eps=0.5, min_samples=2).fit(train30_tensor)
y30_labels = clustering30.labels_
n30_clusters = np.amax(y30_labels)

print(y30_labels)
print(n30_clusters)
```

```python
master0_df['cluster'] = y0_labels
master30_df['cluster'] = y30_labels
```

```python
# TODO: "continue-for" if the last element is reached
def subplot_gen(cluster_df, n_row, n_col, n_fig, path, train_np):
    
    i_c = cluster_df.index.values.astype(int)
    count_c = 0
    l_c = len(cluster_df)
    
    for i in range(n_fig):
        fig, axs = plt.subplots(n_col, n_row, figsize=(25,25), constrained_layout=True,
                        sharex=True, sharey=True)#,sharex='col', sharey='row')
        plt.figure(figsize=(5,5), dpi= 100, facecolor='w', edgecolor='k')
        plt.rcParams['figure.figsize'] = [16, 24]
        plt.rcParams['figure.dpi'] = 100 # 200 e.g. is really fine, but slower
        plt.tight_layout(pad=0.01, w_pad=0, h_pad=0)

        for k in range(n_row):
            for j in range(n_col):
                if count_c < l_c:
                    axs[j, k].imshow(train_np[i_c[count_c]].reshape(144,144),cmap='gray', interpolation='nearest')
                    axs[j, k].axis('off')
                    count_c += 1

        plt.close() #try to remove this
        path_filename = path+str(i)+".png"
        plt.savefig(path_filename)
        plt.close("all")
    
    return
```
